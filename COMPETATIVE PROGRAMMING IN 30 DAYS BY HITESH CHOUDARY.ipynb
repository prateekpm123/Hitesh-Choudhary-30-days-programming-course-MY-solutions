## DAY 4 CHALLENGE

Jack bought 400 hot dogs for the school picnic. If they were contained in packages of 8 hot dogs, how many total packages did he buy ?
Create a python program without using \ or % operator.

```python
stuff_quantity = int(input("Enter the amount of stuff bought\n"))
stuff_per_bag = int(input("Enter the amount of stuff for per bag\n"))

count = 0

while stuff_quantity is not 0:
    stuff_quantity -= stuff_per_bag
    count += 1
    
print("The total number of bags required for distributing the stuff =",count)
```

```python
n = int(input("Enter the number of elements you want in the list\n"))
list = []

# Taking elements in the list from the user

for m in range(0,n):
    apd = int(input("Enter the elements"))
    list.append(apd)
    
print(list)
llist = len(list)
k=0

# Logic for bubble sort

for i in range(llist-1):
    for j in range(llist-1):
        if list[j]>list[j+1]:
            temp = list[j]
            list[j]=list[j+1]
            list[j+1]=temp
            k += 1

print("The sorted list is",list)
```
